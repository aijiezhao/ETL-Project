```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
obese_data = "Obesity.csv"
```

```python
obese_df = pd.read_csv(obese_data)
```

```python
new_df = obese_df.loc[(obese_df['Class'] == 'Obesity / Weight Status')]
```

```python
second_df = new_df.drop_duplicates()
second_df
```

```python
third_df = second_df.loc[(second_df['YearStart'] == 2018)]
```

```python
df = third_df.drop(columns=['YearEnd', 'LocationAbbr', 'Datasource', 'Class', 'Topic', 'Data_Value_Unit', 'Data_Value_Type', 'Data_Value_Alt', 'Data_Value_Footnote_Symbol', 'Data_Value_Footnote', 'Low_Confidence_Limit', 'Total', 'Age(years)', 'Education', 'Gender', 'Income', 'Race/Ethnicity', 'GeoLocation', 'ClassID', 'TopicID', 'QuestionID', 'DataValueTypeID', 'LocationID', 'StratificationCategory1', 'Stratification1', 'StratificationCategoryId1', 'StratificationID1'])
```

```python
#df
new_sorted_df = df.drop(columns=['Sample_Size'])
sorted_df = new_sorted_df.loc[new_sorted_df['Question'] == 'Percent of adults aged 18 years and older who have obesity']
sorted_df
```

```python
final_df = sorted_df.drop(columns=['High_Confidence_Limit '])
```

```python
final_df
```

```python
etl_df = final_df.rename(columns={'YearStart': 'Year', 'LocationDesc': 'state', 'Question': 'question', 'Data_Value': 'data_value'})
```

```python
final_df = etl_df.drop(index=7809, columns=['Year'])
```

```python
final_df
```

```python
final_df = final_df.drop_duplicates(subset = 'state')
```

```python
final_df['data_value'] = pd.to_numeric(final_df['data_value'])
```

```python
final_df.to_csv('obesity_clean.csv', index=False)
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/ETL_Project"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
final_df.to_sql(name='obesity', con=engine, if_exists='append', index=False)
```
